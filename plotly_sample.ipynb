```python
f_name='/content/drive/MyDrive/data/2021/data2021/TestData.csv'

df1=pd.read_csv(f_name)
#df1=df.copy() #エラーを避けるためにdfからdf1へのcopyを明示しておく
df1.columns=['Index','Time','Voltage','T_high','T_low','Res']
df1=df1[['Time','Voltage','T_high', 'T_low','Res']]
```

```python
#第１図
def plot_group1(i):
  if i != 'Voltage':
    fig1.add_trace(go.Scatter(x=df1['Time'], y=df1[i],name=i)
  #  ,mode='markers+lines', marker_size=3, marker_symbol="circle-open", line_width=1)
    )
  else:
    fig1.add_trace(go.Scatter(x=df1['Time'], y=df1[i],name=i)
  #  ,mode='markers+lines', marker_size=3, marker_symbol="circle-open", line_width=1)
    ,secondary_y=True)
  
#第２図
def plot_group2(i):
  fig2.add_trace(go.Scatter(x=df1['Time'], y=df1[i],name=i)
#  ,mode='markers+lines', marker_size=3, marker_symbol="circle-open", line_width=1)
  )


# グラフ表示
fig1= make_subplots(specs=[[{"secondary_y": True}]])
fig2= make_subplots()
[plot_group2(i) if i=='Res' else plot_group1(i) if i !='Time' else _ for i in df1.columns ] #内包表記にしてみた elseの後の'_'は何もしない

fig1.show()
fig2.show()
```
